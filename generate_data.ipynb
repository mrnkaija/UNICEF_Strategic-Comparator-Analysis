```python
import pandas as pd
import numpy as np
np.random.seed(42)

# i generat a synthetic 100k-record SCA dataset across 55 real UNICEF markets

# I create 55 UNICEF fundraising markets
markets = [
    # Africa (15)
    'Kenya', 'Nigeria', 'South Africa', 'Ghana', 'Uganda', 'Ethiopia', 
    'Tanzania', 'Zambia', 'Cote dIvoire', 'Senegal', 'Morocco', 'Egypt', 
    'Algeria', 'Angola', 'Mozambique',
    
    # North America (8)
    'USA-NewYork', 'USA-California', 'USA-Texas', 'USA-Florida', 
    'Canada-Toronto', 'Canada-Montreal', 'Canada-Vancouver', 'Mexico',
    
    # Europe (15)  
    'UK-London', 'UK-Manchester', 'Germany-Berlin', 'Germany-Munich', 
    'France-Paris', 'France-Lyon', 'Italy-Rome', 'Italy-Milan', 
    'Spain-Madrid', 'Spain-Barcelona', 'Netherlands-Amsterdam', 
    'Sweden-Stockholm', 'Norway-Oslo', 'Switzerland-Zurich', 'Belgium-Brussels',
    
    # Asia (10)
    'India-Mumbai', 'India-Delhi', 'Japan-Tokyo', 'China-Shanghai', 
    'SouthKorea-Seoul', 'Australia-Sydney', 'Australia-Melbourne', 
    'Singapore', 'Malaysia-KualaLumpur', 'Philippines-Manila',
    
    # Latin America (5)
    'Brazil-SaoPaulo', 'Brazil-Rio', 'Argentina-BuenosAires', 
    'Chile-Santiago', 'Colombia-Bogota',
    
    # MENA (2)
    'UAE-Dubai', 'SaudiArabia-Riyadh'
]

orgs = ['UNICEF','SaveTheChildren','Oxfam','RedCross','WorldVision','CARE','PlanIntl']
channels = ['PSFR','Digital','Corporate','MajorDonors','Legacy']
regions = ['Africa','North America','Europe','Asia','Latin America','MENA']

data = []
for _ in range(100000):
    market = np.random.choice(markets)
    row = {
        'Market': market,
        'Organization': np.random.choice(orgs),
        'Channel': np.random.choice(channels),
        'Region': dict(zip(markets, ['Africa']*15 + ['North America']*8 + ['Europe']*15 + ['Asia']*10 + ['Latin America']*5 + ['MENA']*2))[market],
        'Income_2024': np.random.uniform(50000, 5000000),
        'Income_2025': 0,  # Will calculate
        'Growth_Rate': np.random.uniform(-0.1, 0.4),
        'Cost_Per_Donor': np.random.uniform(10, 60),
        'Donors': np.random.randint(1000, 50000),
        'Date': pd.Timestamp('2025-01-01') + pd.DateOffset(days=np.random.randint(0,365))
    }
    row['Income_2025'] = row['Income_2024'] * (1 + row['Growth_Rate'])
    data.append(row)

df = pd.DataFrame(data)
df.to_csv('fundraising_data.csv', index=False)
print(f"Generated 100k-record SCA dataset across {len(markets)} real UNICEF markets")
print("\nTop 10 Markets:", df['Market'].value_counts().head(10).to_dict())
print("UNICEF PSFR Growth:", df[(df['Organization']=='UNICEF') & (df['Channel']=='PSFR')]['Growth_Rate'].mean())
```
